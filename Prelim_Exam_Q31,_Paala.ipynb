* Reads an image named "nature.jpg"
* Converts it to grayscale
* Resizes it to 300x300 pixels
* Saves the result as "processed.jpg"

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/CPE 018 Nature/nature.jpg')
```

```python
cv2_imshow(img)
```

```python
converted = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(converted)
```

```python
convert = cv2.imwrite('/content/drive/MyDrive/CPE 018 Nature/converted.jpg', converted)
```

```python
resize = cv2.imread('/content/drive/MyDrive/CPE 018 Nature/converted.jpg')
```

```python
resized = resize[0:300, 0:300]
```

```python
cv2_imshow(resized)
```

```python
cv2.imwrite('/content/drive/MyDrive/CPE 018 Nature/processed.jpg', resized)
```
