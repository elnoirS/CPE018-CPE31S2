```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
circle = cv2.imread('/content/drive/MyDrive/CPE 018 Nature and Circle/circles.png')
```

```python
cv2_imshow(circle)
```

```python
to_gray = cv2.cvtColor(circle, cv2.COLOR_BGR2GRAY)
```

```python
detect_circle = cv2.HoughCircles(to_gray, 3, dp=30, minDist=95, param1=45, param2=60, minRadius=90, maxRadius=100)
```

```python
cv2_imshow(to_gray)
```

```python
detect_circle.shape
```

```python
cv2.imwrite('/content/drive/MyDrive/CPE 018 Nature and Circle/circles_detected.png', detect_circle)
```
